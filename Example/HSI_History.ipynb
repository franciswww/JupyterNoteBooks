```python
data['H1']= (data['High'].shift(-1)).fillna(0)
```

```python
data['H2']= (data['High'].shift(-2)).fillna(0)
```

```python
data['H3']= (data['High'].shift(-3)).fillna(0)
```

```python
data['L1']= (data['Low'].shift(-1)).fillna(0)
```

```python
data['L2']= (data['Low'].shift(-2)).fillna(0)
```

```python
data['L3']= (data['Low'].shift(-3)).fillna(0)
```

```python
data
```

```python
data['Up']= 0
```

```python
data['Up'] = data[['H1','H2','H3']].apply(max, axis=1)/data['Low']
```

```python
data.loc[(data['H1']/data['Low'] > 1.03) | (data['H2']/data['Low'] > 1.03) | (data['H3']/data['Low'] > 1.03), 'Up']=1 
```

```python
data['Down']= 0
```

```python
data.loc[(data['High']/data['L1'] > 1.03) | (data['High']/data['L2'] > 1.03) | (data['High']/data['L3'] > 1.03), 'Down']=1 
```

```python
data['Down'] = data['High']/data[['L1','L2','L3']].apply(min, axis=1)
```

```python
data
```

```python
data.query('Up>=1.03 | Down>=1.03').to_excel("output.xlsx")  
```
